```python
import pandas as pd
import ace_tools as tools;
```

```python
files = ["./COP4520/team15COP4520/output2.csv", "./COP4520/team15COP4520/output3.csv", "./COP4520/team15COP4520/output4.csv"]
dfs = [pd.read_csv(file) for file in files]
finaldf = pd.concat(dfs, ignore_index = True)
```

```python
print(finaldf.head())
```

```python
avg_cracking_time = finaldf.groupby('thread_count')['time_to_crack'].mean()
print(avg_cracking_time)
```

```python
import matplotlib.pyplot as plt

# Plot the average cracking time for each thread count
avg_cracking_time.plot(kind='bar')
plt.xlabel('Thread Count')
plt.ylabel('Average Time to Crack (ns)')
plt.title('Average Cracking Time by Thread Count')
plt.show()
```

```python
avg_cracking_time_2 = dfs[0].groupby('thread_count')['time_to_crack'].mean()
print(avg_cracking_time_2)
avg_cracking_time_2_med = dfs[0].groupby('thread_count')['time_to_crack'].median()
print(avg_cracking_time_2_med)
```

```python
import matplotlib.pyplot as plt

# Plot the average cracking time for each thread count
avg_cracking_time_2.plot(kind='bar')
plt.xlabel('Thread Count')
plt.ylabel('Average Time to Crack (ns)')
plt.title('Average Cracking Time by Thread Count for length 2')
plt.show()

avg_cracking_time_2_med.plot(kind='bar')
plt.xlabel('Thread Count')
plt.ylabel('Average Time to Crack (ns)')
plt.title('Average Cracking Time by Thread Count for length 2')
plt.show()

```
